```
df.columns = [
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'tamanho_empresa'
]
df.head()
```

```
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'MI': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
df['tamanho_empresa'].value_counts()
```

```
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
df['remoto'].value_counts()
```

```
# criando um dataframe do zero para aplicar em uma situação ficctíci ao tratamento de dados em que colocamos infos em valores nulos.
#faremos isso realizando a média de salários

import numpy as np # biblioteca que permite adc valores nulos

df_salarios = pd.DataFrame({
  "nome": ["Ana", "Bruno", "Carlos", "Daniele", 'Val'],
  "salario": [4000, np.nan, 5000, np.nan, 100000]
})

# preenchendo os valores nulos do dataframe com a média dos salários
# salario_media é a coluna nova que tá sendo criada no df_salarios
# = atribui valor a essa nova coluna, nesse caso ta preenchendo com salario.mean q é a média
# round(2) arredonda em 2 números

#calcula a média salarial, subtitui os nulos pela média e arredonda
df_salarios["salario_media"] = df_salarios["salario"].fillna(df_salarios['salario'].mean().round(2))

#calcula a mediana e subtitui os nulos pela média
df_salarios["salario_mediana"] = df_salarios["salario"].fillna(df_salarios['salario'].median())
# a mediana suaviza impactos de informações muito distoantes

df_salarios
```

```
df_temperaturas = pd.DataFrame({
    "Dia": ["seg", "ter", "qua", "qui", "sex"],
    "Temp": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["Preenchido_ffill"] = df_temperaturas["Temp"].ffill()
df_temperaturas
```

```
df_temperaturas = pd.DataFrame({
    "Dia": ["seg", "ter", "qua", "qui", "sex"],
    "Temp": [30, np.nan, np.nan, 28, 27]
})

df_temperaturas["Preenchido_bfill"] = df_temperaturas["Temp"].bfill()
df_temperaturas
```

```
df_cidades = pd.DataFrame({
  "nome": ["Ana", "Bruno", "Carlos", "Daniele", 'Val'],
  "cidade": ["São Paulo", np.nan, "Curitiiba", np.nan, "Belém"]
})

# trocando o nome null em valores nulos por "não informado"
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('não informado')
df_cidades
```

```
import matplotlib.pyplot as plt
```

```
plt.figure(figsize= (8,5))
sns.barplot(data=df_limpo, x='senioridade', y= 'usd')
plt.title("Salário médio por nível de senioridade")
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (USD)')
plt.show() #p mostrar o gráico
```

```
df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = False) # calculando a média dos salários p ficar em orde do maior p menor
```

```
ordem = df_limpo.groupby('senioridade')['usd'].mean().sort_values(ascending = True).index # salva a informação na variável ordem
ordem
```

```
plt.figure(figsize= (8,5))
sns.barplot(data=df_limpo, x='senioridade', y= 'usd', order= ordem)
plt.title("Salário médio por nível de senioridade")
plt.xlabel('Senioridade')
plt.ylabel('Salário médio anual (USD)')
plt.show() #p mostrar o gráic
```

```
import plotly.express as px
```

```
df_media_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reindex(ordem)

fig = px.bar(
    x=df_media_senioridade.index,
    y=df_media_senioridade.values,
    labels={'x': 'Senioridade', 'y': 'Salário médio anual (USD)'},
    title='Salário Médio Anual por Nível de Senioridade (Plotly)'
)
fig.show()
```

```
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade'] #definindo as colunas dessa variável

fig = px.pie(remoto_contagem,
  names = "tipo_trabalho",
  values = 'quantidade',
  title='Proporção dos tipos de trabalho'
)
fig.show()
```
