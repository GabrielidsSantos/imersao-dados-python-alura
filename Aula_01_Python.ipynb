```
import pandas as pd
```

```
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
# lendo uma base de dados usando a biblioteca panda
# usa o read_csv pq o arquivo é csv (separado por vírgulas)
# df é o nome que dei para a base de dados, é muito comum pq significa dataframe
```

```
df.columns = [
    'ano',
    'senioridade',
    'contrato',
    'cargo',
    'salario',
    'moeda',
    'usd',
    'residencia',
    'remoto',
    'empresa',
    'tamanho_empresa'
]
df.head()
```

```
df["senioridade"].value_counts()
# mostra a frequência (quantas vezes) aparece cada info contida na seção "nivel_experiencia"
```

```
df["contrato"].value_counts()
# mostra a frequência (quantas vezes) aparece cada info contida na seção "nivel_experiencia"
```

```
df["remoto"].value_counts()
```

```
df["tamanho_empresa"].value_counts()
```

```
df['senioridade'] = df['senioridade'].replace({
    'SE': 'Sênior',
    'Médio': 'Pleno',
    'EN': 'Júnior',
    'EX': 'Executivo'
})
df['senioridade'].value_counts()
```

```
df['contrato'] = df['contrato'].replace({
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Período',
    'FL': 'Freelancer'
})
df['contrato'].value_counts()
```

```
df['tamanho_empresa'] = df['tamanho_empresa'].replace({
    'M': 'Média',
    'L': 'Grande',
    'S': 'Pequena'
})
df['tamanho_empresa'].value_counts()
```

```
df['remoto'] = df['remoto'].replace({
    0: 'Presencial',
    100: 'Remoto',
    50: 'Híbrido'
})
df['remoto'].value_counts()
```
